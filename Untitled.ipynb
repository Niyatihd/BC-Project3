```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json

import csv
import os
```

```python
CountyLevelDF = pd.read_csv(os.path.join("resources","HistoricalData","CountyLevelSummary.csv"))
CountyLevelDF = CountyLevelDF.drop(["Unnamed: 0"], axis=1)
CountyLevelDF = CountyLevelDF[["Year","County","EstHouseholds","Employees","AvgAnnualPay","Avg.Median Home price"]]


AlamedaDF = CountyLevelDF.loc[CountyLevelDF["County"] =="Alameda"]
AlamedaDF = AlamedaDF.drop(['County'],axis  = 1)
AlamedaDF.set_index("Year",inplace = True)
AlamedaDFchg = AlamedaDF.pct_change()
AlamedaDFchg = AlamedaDFchg.dropna()
AlamedaDFchg["County"] = "Alameda"

ContrCostaDF = CountyLevelDF.loc[CountyLevelDF["County"] =="Contra Costa"]
ContrCostaDF = ContrCostaDF.drop(['County'],axis  = 1)
ContrCostaDF.set_index("Year",inplace = True)
ContrCostaDFchg = ContrCostaDF.pct_change()
ContrCostaDFchg = ContrCostaDFchg.dropna()
ContrCostaDFchg ["County"] = "Contra Costa"
ContrCostaDFchg

MarinDF = CountyLevelDF.loc[CountyLevelDF["County"] =="Marin"]
MarinDF = MarinDF.drop(['County'],axis  = 1)
MarinDF.set_index("Year",inplace = True)
MarinDFchg = MarinDF.pct_change()
MarinDFchg = ContrCostaDFchg.dropna()
MarinDFchg ["County"] = "Marin"
MarinDFchg

NapaDF = CountyLevelDF.loc[CountyLevelDF["County"] =="Napa"]
NapaDF = NapaDF.drop(['County'],axis  = 1)
NapaDF.set_index("Year",inplace = True)
NapaDFchg = NapaDF.pct_change()
NapaDFchg = NapaDFchg.dropna()
NapaDFchg ["County"] = "Marin"
NapaDFchg

SfoDF = CountyLevelDF.loc[CountyLevelDF["County"] =="San Francisco"]
SfoDF = SfoDF.drop(['County'],axis  = 1)
SfoDF.set_index("Year",inplace = True)
SfoDFchg = SfoDF.pct_change()
SfoDFchg = SfoDFchg.dropna()
SfoDFchg ["County"] = "San Francisco"
SfoDFchg

SanMateoDF = CountyLevelDF.loc[CountyLevelDF["County"] =="San Mateo"]
SanMateoDF = SanMateoDF.drop(['County'],axis  = 1)
SanMateoDF.set_index("Year",inplace = True)
SanMateoDFchg = SanMateoDF.pct_change()
SanMateoDFchg = SanMateoDFchg.dropna()
SanMateoDFchg ["County"] = "San Mateo"
SanMateoDFchg

SantaClaraDF = CountyLevelDF.loc[CountyLevelDF["County"] =="Santa Clara"]
SantaClaraDF = SantaClaraDF.drop(['County'],axis  = 1)
SantaClaraDF.set_index("Year",inplace = True)
SantaClaraDFchg = SantaClaraDF.pct_change()
SantaClaraDFchg = SantaClaraDFchg.dropna()
SantaClaraDFchg ["County"] = "San Clara"
SantaClaraDFchg

SonomaDF = CountyLevelDF.loc[CountyLevelDF["County"] =="Sonoma"]
SonomaDF = SonomaDF.drop(['County'],axis  = 1)
SonomaDF.set_index("Year",inplace = True)
SonomaDFchg = SonomaDF.pct_change()
SonomaDFchg = SonomaDFchg.dropna()
SonomaDFchg ["County"] = "Sonoma"
SonomaDFchg

SolanoDF = CountyLevelDF.loc[CountyLevelDF["County"] =="Solano"]
SolanoDF = SolanoDF.drop(['County'],axis  = 1)
SolanoDF.set_index("Year",inplace = True)
SolanoDFchg = SolanoDF.pct_change()
SolanoDFchg = SolanoDFchg.dropna()
SolanoDFchg ["County"] = "Solano"


frames = [SolanoDFchg, SonomaDFchg, SantaClaraDFchg,SanMateoDFchg,SfoDFchg,NapaDFchg,MarinDFchg,ContrCostaDFchg,AlamedaDFchg]
result = pd.concat(frames)
result = result.rename(columns = {"EstHouseholds": "Households","Employees" : "Jobs","AvgAnnualPay" : "Income","Avg.Median Home price": "Median Home Price"})
result.to_csv(os.path.join("resources","HistoricalData","CountyLevelSummary_PctChange.csv"))
result.to_json(orient='table')

result.columns=[''.join(col) for col in result.columns]
result
```
