```python
# Dependencies
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np


import pandas as pd


df = pd.read_csv("resources/HistoricalData/CountyLevelSummary.csv")
df.reset_index(inplace=True, drop=True)
df = df.drop(["Unnamed: 0"], axis=1)
df_County = df.loc[df["County"] =="Alameda"]


x_values = df_County["Year"]
y_values = df_County["AvgAnnualPay"]

slope, intercept, _, _, _ = stats.linregress(x_values, y_values)
fit = slope * x_values + intercept

fig, ax = plt.subplots()

ax.plot(x_values, y_values, linewidth=0, marker='o')
ax.plot(x_values, fit, 'b--')



ax.set_xlabel("Years")
ax.set_ylabel("Annual Pay")

plt.show()
```
