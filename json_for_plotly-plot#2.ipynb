```python
import pandas as pd
import numpy as np
```

```python
data_df = pd.read_csv("static/resources/MinimumQualifyingIncomePastANDFuture_5YrIntervals.csv")
```

```python
data_df.Year = data_df.Year.astype(float)
data_df["Avg.Median Home price"] = data_df["Avg.Median Home price"].astype(float)
data_df["QualifyingIncome"] = data_df["QualifyingIncome"].astype(float)
data_df["AvgAnnualIncome"] = data_df["AvgAnnualIncome"].astype(float)
```

```python
counties = ['Bay Area', 'Solano', 'Sonoma', 'Santa Clara', 'San Mateo',
       'San Francisco', 'Napa', 'Marin', 'Contra Costa', 'Alameda']
final_dict = {}
for county in counties:
    temp_dict = {}
#     print(county)
    current_df = data_df.loc[data_df["County"] == county]
    year_list = current_df['Year'].tolist()       
    avg_income_list = current_df['AvgAnnualIncome'].tolist() 
    qualifying_income_list = current_df['QualifyingIncome'].tolist() 
    homeprice_list = current_df['Avg.Median Home price'].tolist()
    temp_dict["year"] = year_list
    temp_dict["avgAnnualIncome"] = avg_income_list
    temp_dict["qualifyingIncome"] = qualifying_income_list
    temp_dict["medianHomePrice"] = homeprice_list
    final_dict[county] = temp_dict
    print(temp_dict)
# final_dict
```
